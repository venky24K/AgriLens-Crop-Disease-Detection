# 04 FAST Best Quantum Algorithm (Top 10 Classes)

**Hybrid VQC with Data Re-uploading — Optimized for Speed & Accuracy**

Target: 10 most common diseases only (10 classes instead of 101).
Data: 400 images per class (robust sample).
Model: Advanced 6-layer Data Re-uploading circuit.

```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))
import torch
from utils.config import CFG
from data.loader import load_dataset
from data.preprocess import build_dataloaders
from training.train import train_hybrid
from training.evaluate import evaluate_model, plot_training_curves, benchmark_report
from models.classical import load_svm, evaluate_svm, load_rf, evaluate_rf, load_lr, evaluate_lr

classes = ['tomato_tomato_yellowleaf_curl_virus', 'tomato_bacterial_spot', 'potato_early_blight', 'tomato_late_blight', 'apple_apple_scab', 'grape_black_rot', 'apple_cedar_apple_rust', 'tomato_spider_mites_two_spotted_spider_mite', 'tomato_target_spot', 'grape_esca_black_measles']
CFG.CLASSES = classes
CFG.NUM_CLASSES = len(classes)
print(f'Training on {len(classes)} classes...')
print(f'Device: {CFG.DEVICE}')
```

```python
print('Loading Top 10 dataset (400 images per class)...')
train_rec, val_rec, test_rec = load_dataset(max_per_class=400)
train_loader, val_loader, test_loader = build_dataloaders(
    train_rec, val_rec, test_rec,
    batch_size=CFG.VQC_BATCH_SIZE, num_workers=0
)
print(f'Batches -> Train: {len(train_loader)}, Val: {len(val_loader)}, Test: {len(test_loader)}')
```

```python
print('Training Advanced Hybrid VQC (Top 10)...')
model, history = train_hybrid(train_loader, val_loader, epochs=10)
plot_training_curves(history, CFG.PLOTS_DIR / 'hybrid_vqc_fast_curves.png')
print('Training complete!')
```

```python
print('Evaluating on test set...')
results = evaluate_model(model, test_loader, 'Fast Hybrid VQC')
print(f'\nAccuracy: {results["accuracy"]:.4f}')
```
