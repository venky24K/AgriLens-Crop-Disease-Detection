# 02 Classical Baseline (Refined 80k+ Images)

```python
import sys
import os
sys.path.insert(0, os.path.abspath('../src'))

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score

from utils.config import CFG
from data.loader import load_dataset
from data.preprocess import build_feature_matrix
from models.classical import (
    build_svm, train_svm, evaluate_svm, save_svm,
    build_rf, train_rf, evaluate_rf, save_rf,
    build_lr, train_lr, evaluate_lr, save_lr
)

# Force load of the FULL dataset (80k+ images)
print("Loading FULL dataset records from MasterDataset...")
train_rec, val_rec, test_rec = load_dataset(max_per_class=None)
print(f"Records loaded -> Train: {len(train_rec)}, Val: {len(val_rec)}, Test: {len(test_rec)}")

# Extract features centrally
print("Extracting features (this will take a while for 80k+ images)...")
X_train, y_train = build_feature_matrix(train_rec, desc='Train')
X_val, y_val = build_feature_matrix(val_rec, desc='Val')
X_test, y_test = build_feature_matrix(test_rec, desc='Test')
print(f'Feature Matrices Shape -> Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}')
```

```python
print('Training SVM baseline...')
svm_model = train_svm(X_train, y_train)
save_svm(svm_model)
svm_results = evaluate_svm(svm_model, X_test, y_test)
```

```python
print('Training Random Forest baseline...')
rf_model = train_rf(X_train, y_train)
save_rf(rf_model)
rf_results = evaluate_rf(rf_model, X_test, y_test)
```

```python
print('Training Logistic Regression baseline...')
lr_model = train_lr(X_train, y_train)
save_lr(lr_model)
lr_results = evaluate_lr(lr_model, X_test, y_test)
```

```python
from training.evaluate import benchmark_report
# In this notebook, we compare SVM vs others. 
# We pass SVM as 'hybrid_results' to satisfy the signature since we don't have VQC here yet.
benchmark_report(
    hybrid_results=svm_results, 
    cnn_results=None, 
    svm_results=svm_results, 
    rf_results=rf_results, 
    lr_results=lr_results
)
```
